```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
df_boston = pd.DataFrame(data = boston.data, columns = boston.feature_names)
df_boston.shape
```

```python
df_boston.head()
```

```python
df_boston.describe()
```

```python
df_boston["Price"] = boston.target
df_boston
```

```python
df_boston.shape
```

```python
df_boston.info()
```

```python
x_data = df_boston[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
y_data = df_boston['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split(x_data,y_data,test_size=0.2,random_state=0)
print(f"x train shape = {X_train.shape}")
print(f"x test shape = {X_test.shape}")
print(f"y train = {y_train.shape}")
print(f"y test = {y_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.score(X_train, y_train)
y_pred 
```

```python
from sklearn.metrics import r2_score

r2_pred = regressor.predict(X_train)
r2_predic = r2_score(y_train,r2_pred)
r2_predic
```

```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python

df = df['mean area']
```

```python
nump = df.to_numpy()
```

```python
nump
```

```python
def binarySearch(array, x, low, high):
    while low <= high:
        mid = low+ (high-low)//2
        if array[mid]==x:
            return mid
        elif array[mid]< x:
            low = mid+ 1
        else :
            high = mid-1
    return -1

x = 807.2
sortNum = sorted(nump)
lensor = len(sortNum)
result = binarySearch(sortNum, x, 0, lensor-1)
if result !=-1:
    print(f"element found at index {result} with value ")
    
else:
    print("element not found")
```

```python
import time
awal = time.time()
def linearSearch(array, n, x):
    for i in range(0,n):
        if(array[i]==x):
            return i
    return -1

x = 807.2
n = len(nump)
result = linearSearch(nump, n, x)
akhir = time.time()
if (result ==-1):
    print("element not found")
else:
    print(f"element found at index {result}")

print(f"waktu eksekusi = {awal-akhir}")
```
